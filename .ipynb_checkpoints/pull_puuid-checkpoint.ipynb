```python
import requests, urllib.parse

# ðŸ§  Replace this with your current valid Riot API key
API_KEY = "RGAPI-e0b5bf3f-4ef3-4f55-b791-1e0a9e6fc472".strip()

# Faker's Riot ID
game_name = "Hide on bush"
tag_line  = "KR1"

# The account endpoint for KR pros is under the 'asia' routing cluster
region = "asia"

# Build the request URL (Account-V1)
url = f"https://{region}.api.riotgames.com/riot/account/v1/accounts/by-riot-id/{urllib.parse.quote(game_name)}/{urllib.parse.quote(tag_line)}"

# Send the request
response = requests.get(url, headers={"X-Riot-Token": API_KEY})

# Handle the response
if response.status_code == 200:
    data = response.json()
    print("Game Name:", data["gameName"])
    print("Tag Line:", data["tagLine"])
    print("PUUID:", data["puuid"])
else:
    print("Error:", response.status_code)
    print("Details:", response.text)
```
